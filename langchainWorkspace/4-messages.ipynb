```python
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
```

```python
from langchain.messages import HumanMessage, SystemMessage, AIMessage, ToolMessage

messages = [
    SystemMessage(content="You are a helpful assistant that translates English to French."),
    HumanMessage(content="Translate the following sentence to French: 'Hello, how are you?'"),
] 
response = model.invoke(messages)
print(response.content)
```

```python
response.usage_metadata
```

```python
### Messages with tools
ai_message = AIMessage(
    content = [],
    tool_calls = [{
        "name": "get_current_weather",
        "args": {"location": "New York"},
        "id": "call_123"
    }]
)

### Execute tool and get the result
weather_result = "The current weather in New York is sunny."
tool_message = ToolMessage(content=weather_result, tool_call_id = "call_123")

messages = [
    SystemMessage(content="You are a helpful assistant that predicts the weather."),
    HumanMessage(content="What's the weather like in New York?"),
    ai_message,     # Model's tool call to get_current_weather
    tool_message,   # Tool execution result from get_current_weather
]

response = model.invoke(messages)
response
```
