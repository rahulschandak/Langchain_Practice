## Stream

Without Streaming: The server waits for the AI to finish generating the entire response (all tokens). Only when it's 100% complete does it package it up and send it back to your app.

With Streaming (Server-Sent Events): Instead of waiting, the server opens a continuous connection. As soon as the AI generates a single token, the server immediately shoots it to your app.

```python
for chunk in model.stream("Write me a short paragraph on artificial intelligence"):
    print(chunk.text, end='|', flush = True)
```

```python
for chunk in model.stream("Why do parrots have different feather colors"):
    print(chunk.text, end='|', flush = True)
```

## Batch

Batch refers to grouping multiple independent inputs or tasks to be processed by the model concurrently rather than one-by-one. This technique optimizes hardware throughput, reduces network overhead, and improves efficiency.

Example: 
Imagine you have an AI agent that needs to translate 100 customer support tickets.

Sequential: The agent translates ticket 1, waits for the response, translates ticket 2, waits, and so on (taking a long time).

Batched: The agent groups all 100 tickets into a single list and sends them to the model to be processed in parallel at the same time.

```python
responses = model.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is AI?"
])
for response in responses:
    print(response)
```

```python
responses = model.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is AI?",
    "How many days are there in a year?"
], config = {
    'max_concurrency' : 3,  # Limit to 3 parallel calls
})
for response in responses:
    print(response)
```
