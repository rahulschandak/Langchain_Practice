## Tools in AI Agents / LangChain

**Tools** are interfaces that an agent or LLM can use to interact with the outside world. LLMs on their own can only generate text based on their training data. Tools allow them to:
- Search the web
- Query databases
- Perform complex calculations
- Call external APIs

### How Tool Calling Works:
1. **Define the Tool**: You create a function in Python and describe what it does and what arguments it expects.
2. **Bind the Tool to the Model**: You register the tool with the LLM.
3. **Ask the Model**: When you ask a query like "What is 357 multiplied by 12?", the model realizes it has a multiplication tool, stops writing text, and outputs a request to call that tool with `{'a': 357, 'b': 12}`.
4. **Execute the Tool**: Your application code runs the actual Python function and passes the result back to the model to finalize the answer.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")

model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("Why do parrots fly?")
response.content
```

```python
from langchain.tools import tool

@tool
def get_current_weather(location: str) -> str:
    """Get the current weather in a given location"""
    return f"The current weather in {location} is sunny."

model_with_tools = model.bind_tools([get_current_weather])
```

```python
response = model_with_tools.invoke("What is the current weather in New York?")  
print(response)
for tool_call in response.tool_calls:
    print(f"Tool called: {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
```
